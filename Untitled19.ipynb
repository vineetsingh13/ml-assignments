```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.drop(['TOTAL COST TO HOSPITAL ', 'SL.','GENDER','MARITAL STATUS', 'KEY COMPLAINTS -CODE'], axis=1)  # Removing non-numeric and target columns
y = df['TOTAL COST TO HOSPITAL ']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

independent_variables = ['SL.', 'AGE', 'BODY WEIGHT', 'BODY HEIGHT', 'HR PULSE', 'RR']

scaler = StandardScaler()
df_normalized = scaler.fit_transform(df[independent_variables])
df_normalized = pd.DataFrame(df_normalized, columns=independent_variables)

# List of columns that need one-hot encoding
columns_to_encode = ['GENDER', 'MARITAL STATUS', 'KEY COMPLAINTS -CODE']

# Apply one-hot encoding to the dataset
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_cols = encoder.fit_transform(df[columns_to_encode])

# Get the column names for the encoded features
encoded_column_names = encoder.get_feature_names_out(columns_to_encode)

# Create a DataFrame for the encoded features
encoded_df = pd.DataFrame(encoded_cols, columns=encoded_column_names)

# Concatenate the encoded columns with the original dataset
df_encoded = pd.concat([df.drop(columns=columns_to_encode), encoded_df], axis=1)

# Update the list of independent variables


# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df_normalized, df[dependent_variable], test_size=0.2, random_state=42)


# Iterate through each column
for column in independent_variables:
    # Extracting the current column for training and testing
    X_train_col = X_train[[column]]
    X_test_col = X_test[[column]]

    # Creating a linear regression model
    model = LinearRegression()

    # Fitting the model
    model.fit(X_train_col, y_train)

    # Making predictions
    y_pred = model.predict(X_test_col)

    # Calculating Mean Squared Error (MSE)
    mse = mean_squared_error(y_test, y_pred)

    # Calculating R-squared score
    r2 = r2_score(y_test, y_pred)

    # Printing results
    print(f"Column: {column}")
    print(f"MSE: {mse}")
    print(f"R-squared score: {r2}")
    print("---------------------")
```

```python
for i in range(len(independent_variables)):
    for j in range(i+1, len(independent_variables)):
        # Extracting the current pair of columns for training and testing
        X_train_pair = X_train[[independent_variables[i], independent_variables[j]]]
        X_test_pair = X_test[[independent_variables[i], independent_variables[j]]]

        # Creating a linear regression model
        model = LinearRegression()

        # Fitting the model
        model.fit(X_train_pair, y_train)

        # Making predictions
        y_pred = model.predict(X_test_pair)

        # Calculating Mean Squared Error (MSE)
        mse = mean_squared_error(y_test, y_pred)

        # Calculating R-squared score
        r2 = r2_score(y_test, y_pred)

        # Printing results
        print(f"Columns: {independent_variables[i]}, {independent_variables[j]}")
        print(f"MSE: {mse}")
        print(f"R-squared score: {r2}")
        print("---------------------")
```
