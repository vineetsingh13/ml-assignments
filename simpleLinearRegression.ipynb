```python
df=pd.read_csv("FuelConsumptionCo2.csv")
```

```python
cdf=df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head()
```

```python
viz=cdf[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
viz.hist()
plt.show()
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB,cdf.CO2EMISSIONS,color='blue')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("CO2EMISSIONS")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS,cdf.CO2EMISSIONS,color='blue')
plt.xlabel("cylinders")
plt.ylabel("engine size")
plt.show()
```

```python
#here we define a mask of random number which is less than 0.8 and based on that random number we divide the data into train/test split
#Let's split our dataset into train and test sets. 80% of the entire dataset will be used for training and 20% for testing. We create a mask to select random rows using np.random.rand() function:

msk=np.random.rand(len(df)) < 0.8
train=cdf[msk]
test=cdf[~msk]
```

```python
plt.scatter(train.ENGINESIZE,train.CO2EMISSIONS,color='blue')
plt.xlabel("engine size")
plt.ylabel("co2 emissions")
plt.show()
```

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()

#so here we first convert the data into an array uing asanyarray()
#we use asanyarray() because if the data is not an ndimen array it converts it into an ndimen array if it already is it doesnot do anything

train_x=np.asanyarray(train[['ENGINESIZE']])
train_y=np.asanyarray(train[['CO2EMISSIONS']])

#here we use the fit function to fit the best regression line based on training data
regr.fit(train_x,train_y)

print("coefficient",regr.coef_)
print("intercept",regr.intercept_)
```

```python
plt.scatter(train.ENGINESIZE,train.CO2EMISSIONS,color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("engine size")
plt.ylabel("co2 emissions")
plt.show()
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])

#here we predict the dependent variable test_y values using independent variable test_x
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```

```python
#lets check for another independent variables

train_x1=np.asanyarray(train[['CYLINDERS']])
train_y1=np.asanyarray(train[['CO2EMISSIONS']])

regr.fit(train_x1,train_y1)

print("coefficient",regr.coef_)
print("intercept",regr.intercept_)
```

```python
plt.scatter(cdf.CYLINDERS,cdf.CO2EMISSIONS,color='blue')
plt.plot(train_x1, regr.coef_[0][0]*train_x1 + regr.intercept_[0], '-r')
plt.xlabel("cylinders")
plt.ylabel("co2 emissions")
plt.show()
```
