```python
df.dropna(inplace=True)
```

```python
X=df.drop(columns=["MEDV"])
Y=df["MEDV"]
```

```python

Y.head()
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
print("training set ", X_train.shape, y_train.shape)
print("testing set: ",X_test.shape,y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
regression_tree = DecisionTreeRegressor(criterion = 'squared_error')
```

```python
regression_tree.fit(X_train,y_train)
```

```python
regression_tree.score(X_test,y_test)
```

```python
prediction=regression_tree.predict(X_test)
print("$",(prediction - y_test).abs().mean()*1000)
```

```python
regression2=DecisionTreeRegressor(criterion='friedman_mse')
regression2.fit(X_train,y_train)
regression2.score(X_test,y_test)

pred=regression2.predict(X_test)
print("$",(pred - y_test).abs().mean()*1000)
```

```python
regression3=DecisionTreeRegressor(criterion='absolute_error')
regression3.fit(X_train,y_train)
regression3.score(X_test,y_test)

pred1=regression3.predict(X_test)
print("$",(pred1 - y_test).abs().mean()*1000)
```
