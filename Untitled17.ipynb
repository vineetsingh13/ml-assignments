```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("hospital_data.csv")
df
```

```python
df2 = pd.get_dummies(df, columns=['GENDER', 'MARITAL STATUS', 'KEY COMPLAINTS -CODE'], drop_first=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression


correlation_matrix = df2.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title("Correlation Matrix")
plt.show()

# Selecting features using SelectKBest with f_regression
X = df2.drop(['TOTAL COST TO HOSPITAL ', 'SL.','GENDER','MARITAL STATUS', 'KEY COMPLAINTS -CODE'], axis=1)  # Removing non-numeric and target columns
y = df2['TOTAL COST TO HOSPITAL ']
selector = SelectKBest(score_func=f_regression, k='all')
selector.fit(X, y)
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': selector.scores_})
feature_scores.sort_values(by='Score', ascending=False, inplace=True)

# Plotting feature scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Feature', data=feature_scores)
plt.title('Feature Importance Scores')
plt.xlabel('Score')
plt.ylabel('Feature')
plt.show()

# Print top features
print("Top Features:")
print(feature_scores)
```

```python
top_features = feature_scores['Feature'].tolist()[:5]
```

```python
top_features
```

```python
X_top = X[top_features]
```

```python
X_top
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train_top, X_test_top, y_train, y_test = train_test_split(X_top, y, test_size=0.2, random_state=42)

# Fit the linear regression model using the top features
model_top = LinearRegression()
model_top.fit(X_train_top, y_train)

# Make predictions on the test set using the top features
y_pred_top = model_top.predict(X_test_top)

# Evaluate the model using the top features
mse_top = mean_squared_error(y_test, y_pred_top)
r2_top = r2_score(y_test, y_pred_top)
print("Mean Squared Error (Top Features):", mse_top)
print("R^2 Score (Top Features):", r2_top)
```
