```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
df=pd.read_csv("Iris.csv")
df
```

```python
df['Species'].unique()
```

```python
df.drop(columns="Id",inplace=True)
df
```

```python
df.isnull().sum()
```

```python
import seaborn as sns

plt.subplots(figsize = (8,8))
sns.heatmap(df.corr(),annot=True,fmt=".5f",cmap='coolwarm')
plt.show()
```

```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df['Species']
```

```python
X=np.asanyarray(X)
X
```

```python
y=np.asanyarray(y)
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
print("training set", X_train.shape,y_train.shape)
print("testing set", X_test.shape,y_test.shape)
```

```python
from sklearn.metrics import accuracy_score

nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print('accuracy_Naive Bayes: %.3f' %accuracy)
```

```python
df
```

```python
y_vals, y_ints = np.unique(y, return_inverse=True)
print(y_vals,y_ints)
```

```python
y_cats = to_categorical(y_ints)
y_cats
```

```python
X_train_nn,X_test_nn,y_train_nn,y_test_nn=train_test_split(X,y_cats,test_size=0.3,random_state=0)
print("training set", X_train_nn.shape,y_train_nn.shape)
print("testing set", X_test_nn.shape,y_test_nn.shape)
```

```python
model=Sequential()
model.add(Dense(16, input_shape=(4,),activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

model.fit(X_train_nn, y_train_nn, epochs=100, batch_size=1, verbose=1)
```

```python
loss, accuracy_nn = model.evaluate(X_test_nn, y_test_nn, verbose=1)
print("\n\nLoss: %6.4f\tAccuracy: %6.4f" % (loss, accuracy_nn))
```

```python
algo=['naive bayes','neural network']
acc=[accuracy,accuracy_nn]

plt.bar(algo, acc, color=['blue', 'green'])
plt.xlabel('Algorithms')
plt.ylabel('Accuracy')
plt.title('Accuracy of Two Algorithms')
plt.show()
```
